```python
#Importing Relevant Libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Reading the Data
df = pd.read_excel(r'Samples for Python.xlsx', sheet_name='Data Sample')
df

#Creating the Pie Chart
df["Segment"].value_counts().plot(kind='pie', fig=plt.figure(figsize =(10,10)), autopct='%1.0f%%')
shadow=True
ax = plt.gca()
ax.get_yaxis().set_visible(False)
plt.title("Percentage of Sales per Region", fontweight='bold')
plt.legend (loc="right", bbox_to_anchor =(0.2,0))
plt.rcParams.update({'font.size': 20})

#Creating the Table
table=pd.crosstab(index=df['Region'], columns = 'Region').apply(lambda r: r/len(df), axis=1)
table.style.set_table_styles([{'selector' : '','props' : [('border','2px solid black')]}])
```
