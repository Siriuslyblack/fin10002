```python
# imports the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# loading the dataset
variable_name = pd.read_excel (r'Samples for Python.xlsx', sheet_name='Data Sample')

# set a title
plt.title('Total Sales per Month', fontsize = 12)

# label x and y axis
plt.xlabel ('Order Date')
plt.ylabel ('Total Sales')

# plot the graph
plt.plot(variable_name ["Order Date"], variable_name ["Sales"])
plt.xticks(rotation=30)

# print the graph
plt.show()


#Import module for Table generation
from tabulate import tabulate

# Import Excel data
import pandas as pd
data = pd.read_excel (r'Samples for Python.xlsx', sheet_name='Data Sample')
df = pd.DataFrame (data, columns=['Months','Total Sales $'])
# Print table
print(tabulate(data, headers= df))
```
